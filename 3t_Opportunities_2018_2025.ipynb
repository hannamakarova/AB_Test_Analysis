```python
import pandas as pd
import numpy as np


from google.colab import drive
drive.mount("/content/drive")

# Change directory to your specific folder
%cd /content/drive/MyDrive/3t

# Read the Excel file
data = pd.read_excel("3t_All_Opportunities_2018-2025.xlsx")

# Display the first few rows
data.head()
```

```python
duplicates = data[data.duplicated()]
print(duplicates)
```

```python
data.drop_duplicates()
```

```python
data['Opportunity Number'] = pd.to_numeric(data['Opportunity Number'], errors='coerce')
data.info()
```

```python
data['3t Division'] = data['3t Division'].replace("3t - Digital", "3t Digital")
data['3t Division'] = data['3t Division'].replace("3t - EnerMech", "3t EnerMech")
data['3t Division'] = data['3t Division'].replace("3t - GTSC", "3t GTSC")
data['3t Division'] = data['3t Division'].replace("3t - ALL STOP!", "3t ALL STOP")

data['Opportunity Sector'] = data['Opportunity Sector'].replace("Oil and Gas – Downstream", "Oil and Gas - Downstream")
data['3t Department']= data['3t Division'].str.split("-").str[0]
data['3t Department Type']= data['3t Division'].str.split("-").str[1]
data['Sector']= data['Opportunity Sector'].str.split("-").str[0]
data['Sector Type']= data['Opportunity Sector'].str.split("-").str[1]

data['3t Department Type'].unique()
```

```python
data_clean= data[~((data["3t Division"].isna()) & (data["Opportunity Sector"].isna()) & (data["Status"] == "Lost"))]

data_clean.info()
```

```python
data_clean['Opportunity Sector'].fillna('Unknown', inplace=True)
data_clean['Country'].fillna('Unknown', inplace=True)
data_clean['Client'].fillna('Unknown', inplace=True)
data_clean['3t Division'].fillna('Unknown', inplace=True)
data_clean['Opportunity Number'].bfill()

data_clean.isna().sum()

```
